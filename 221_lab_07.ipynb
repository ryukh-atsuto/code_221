```python
import sys
import heapq

def dijkstra(N, graph, start):
    dist = [float('inf')] * (N + 1)
    dist[start] = 0
    pq = [(0, start)]
    while pq:
        current_dist, current_node = heapq.heappop(pq)

        if current_dist > dist[current_node]:
            continue

        for neighbor, weight in graph[current_node]:
            distance = current_dist + weight

            if distance < dist[neighbor]:
                dist[neighbor] = distance
                heapq.heappush(pq, (distance, neighbor))

    return dist

def find_meeting_point(N, M, S, T, edges):

    graph = [[] for _ in range(N + 1)]
    for u, v, w in edges:
        graph[u].append((v, w))

    dist_from_S = dijkstra(N, graph, S)
    dist_from_T = dijkstra(N, graph, T)

    min_time = float('inf')
    meeting_node = -1

    for v in range(1, N + 1):
        if dist_from_S[v] < float('inf') and dist_from_T[v] < float('inf'):
            time_to_meet = max(dist_from_S[v], dist_from_T[v])
            if time_to_meet < min_time:
                min_time = time_to_meet
                meeting_node = v
            elif time_to_meet == min_time:
                meeting_node = min(meeting_node, v)

    if meeting_node == -1:
        print(-1)
    else:
        print(min_time, meeting_node)


input = sys.stdin.read
data = input().splitlines()
N, M, S, T = map(int, data[0].split())
edges = [tuple(map(int, line.split())) for line in data[1:M + 1]]
find_meeting_point(N, M, S, T, edges)
```

```python
import sys
import heapq

def min_danger_levels(N, M, roads):

    graph = [[] for _ in range(N + 1)]
    for u, v, w in roads:
        graph[u].append((v, w))
        graph[v].append((u, w))

    dist = [float('inf')] * (N + 1)
    dist[1] = 0


    pq = [(0, 1)]

    while pq:
        current_danger, current_city = heapq.heappop(pq)

        if current_danger > dist[current_city]:
            continue


        for neighbor, danger in graph[current_city]:

            max_danger = max(current_danger, danger)

            if max_danger < dist[neighbor]:
                dist[neighbor] = max_danger
                heapq.heappush(pq, (max_danger, neighbor))

    result = []
    for i in range(1, N + 1):
        if dist[i] == float('inf'):
            result.append(-1)
        else:
            result.append(dist[i])

    return result

input = sys.stdin.read
data = input().splitlines()
N, M = map(int, data[0].split())
roads = [tuple(map(int, line.split())) for line in data[1:M + 1]]

result = min_danger_levels(N, M, roads)

print(' '.join(map(str, result)))
```

```python
import sys
import heapq

def min_cost_path(N, M, S, D, weights, edges):

    graph = [[] for _ in range(N + 1)]
    for u, v in edges:
        graph[u].append(v)

    cost = [float('inf')] * (N + 1)
    cost[S] = weights[S - 1]

    pq = [(cost[S], S)]

    while pq:
        current_cost, current_node = heapq.heappop(pq)

        if current_node == D:
            return current_cost

        if current_cost > cost[current_node]:
            continue


        for neighbor in graph[current_node]:
            new_cost = current_cost + weights[neighbor - 1]


            if new_cost < cost[neighbor]:
                cost[neighbor] = new_cost
                heapq.heappush(pq, (new_cost, neighbor))

    return -1 if cost[D] == float('inf') else cost[D]

input = sys.stdin.read
data = input().splitlines()
N, M, S, D = map(int, data[0].split())
weights = list(map(int, data[1].split()))
edges = {tuple(map(int, line.split())) for line in data[2:M + 2] }

result = min_cost_path(N, M, S, D, weights, edges)
print(result)
```

```python
import sys
import heapq

def second_shortest_path(N, M, S, D, edges):

    graph = [[] for _ in range(N + 1)]
    for u, v, w in edges:
        graph[u].append((v, w))
        graph[v].append((u, w))
    dist = [[float('inf')] * 2 for _ in range(N + 1)]
    dist[S][0] = 0

    pq = [(0, S)]

    while pq:
        current_dist, current_node = heapq.heappop(pq)

        if current_dist > dist[current_node][1]:
            continue


        for neighbor, weight in graph[current_node]:
            new_dist = current_dist + weight


            if new_dist < dist[neighbor][0]:
                dist[neighbor][1] = dist[neighbor][0]
                dist[neighbor][0] = new_dist
                heapq.heappush(pq, (new_dist, neighbor))

            elif dist[neighbor][0] < new_dist < dist[neighbor][1]:
                dist[neighbor][1] = new_dist
                heapq.heappush(pq, (new_dist, neighbor))

    second_shortest = dist[D][1]
    return second_shortest if second_shortest != float('inf') else -1

input = sys.stdin.read
data = input().splitlines()
N, M, S, D = map(int, data[0].split())
edges = [tuple(map(int, line.split())) for line in data[1:M + 1]]

result = second_shortest_path(N, M, S, D, edges)

print(result)
```
