```python
def bubble_sort(ids, marks):
    n = len(marks)
    swap_count = 0

    for i in range(n - 1):
        for j in range(n - i - 1):
            if marks[j] < marks[j + 1] or (marks[j] == marks[j + 1] and ids[j] > ids[j + 1]):  # Sorting in descending order with ID priority
                marks[j], marks[j + 1] = marks[j + 1], marks[j]
                ids[j], ids[j + 1] = ids[j + 1], ids[j]
                swap_count += 1

    return swap_count, ids, marks

# Input data
n = 7
ids = [7, 4, 9, 3, 2, 5, 1]
marks = [40, 50, 50, 20, 10, 10, 10]

# Sorting using bubble sort
swap_count, sorted_ids, sorted_marks = bubble_sort(ids, marks)

# Output results
print(f"Minimum swaps: {swap_count}")
for i in range(n):
    print(f"ID: {sorted_ids[i]} Mark: {sorted_marks[i]}")

```

```python
def MARKSort(arr):
    count = 0
    n = len(arr)

    for i in range(n - 1):
        max_idx = i
        for j in range(i + 1, n):
            # Sorting primarily by marks (descending) and secondarily by ID (ascending)
            if arr[j][1] > arr[max_idx][1] or (arr[j][1] == arr[max_idx][1] and arr[j][0] < arr[max_idx][0]):
                max_idx = j

        if max_idx != i:
            arr[i], arr[max_idx] = arr[max_idx], arr[i]
            count += 1  # Increment swap count

    return count

# Input data
loop_count = int(input())

# Constraints Check
if not (1 <= loop_count <= 1000):
    exit()

list_student = []
ids = list(map(int, input().split()))
marks = list(map(int, input().split()))

for i in range(loop_count):
    if not (1 <= ids[i] <= 1000 and 1 <= marks[i] <= 1000):
        exit()
    list_student.append((ids[i], marks[i]))

# Sorting & Counting Swaps
swapped = MARKSort(list_student)

# Output Results
print(f"Minimum swaps: {swapped}")
for student in list_student:
    print(f"ID: {student[0]} Mark: {student[1]}")
```
