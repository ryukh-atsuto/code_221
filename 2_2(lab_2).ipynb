```python
def Matched(arr,counter):
  s=0
  e=len(arr)-1
  while s<e:
    if arr[s]+arr[e]==counter:
      print(f"{s+1} {e+1}")
      return
    elif arr[s]+arr[e]> counter:
      e=e-1
    elif arr[s]+arr[e]< counter:
      s=s+1
  print(-1)
  return

digit, counter = map(int, input().split())
arr=[]
elem=input()
arr_list=elem.split()
for i in range (digit):
  arr.append(int(arr_list[i]))
Matched(arr,counter)
```

```python
def Beautiful_Sorted_List(arr1,arr2):
  sort_arr=[]
  indx_1=0
  indx_2=0
  while indx_1<len(arr1) or indx_2<len(arr2):
    if indx_1<len(arr1) and indx_2<len(arr2):
      if arr1[indx_1]<arr2[indx_2]:
        sort_arr.append(arr1[indx_1])
        indx_1+=1
      else:
        sort_arr.append(arr2[indx_2])
        indx_2+=1
    elif indx_1>=len(arr1) and indx_2<len(arr2):
         sort_arr.append(arr2[indx_2])
         indx_2+=1
    elif indx_2>=len(arr2) and indx_1<len(arr1):
         sort_arr.append(arr1[indx_1])
         indx_1+=1
  return sort_arr


digit_1 = int(input())
arr1 = list(map(int, input().split()))
digit_2 = int(input())
arr2 = list(map(int,input().split()))
Ans = Beautiful_Sorted_List(arr1,arr2)
print(" ".join(map(str,Ans)))
```

```python
import sys

def find_first1(string):
    st, end = 0, len(string) - 1
    first_one = -1

    while st <= end:
        mid = (st + end) // 2
        if string[mid] == "1":
            first_one = mid + 1  # Convert to 1-based index
            end = mid - 1  # Search left half
        else:
            st = mid + 1  # Search right half

    return first_one if first_one != -1 else -1

def solve():
    # Read input efficiently
    input_data = sys.stdin.read().split()
    T = int(input_data[0])
    results = [str(find_first1(input_data[i])) for i in range(1, T + 1)]

    # Print output efficiently
    sys.stdout.write("\n".join(results) + "\n")
```

```python
def bs_left(arr,val):
    st = 0
    end = len(arr) -1
    while st < end:
        mid = (st+end) //2
        if val <= arr[mid]:
            end = mid
        else:
            st = mid + 1
    return st

def bs_right(arr,val):
    st = 0
    end = len(arr)-1
    while st < end:
        mid = (st+end)//2
        if val < arr[mid]:
            end = mid
        else:
            st = mid + 1
    if arr[st] <= val:
        return st+1
    return st
def countNumbers(arr,query):
    return(bs_right(arr,query[1])-bs_left(arr,query[0]))
Array, compare = map(int, input().split())
arr = list(map(int,input().split()))
for _ in range(compare):
    query = list(map(int,input().split()))
    print(countNumbers(arr,query))
```
