```python
from collections import deque
def maxNumber(graph,n):
    team = [None]*(n+1)
    count = 0

    for i in range(1,n+1):
        if team[i] == None:
            robots = 1
            humans = 0
            team[i] = 0
            queue = deque()
            queue.append(i)
            while queue:
                u = queue.popleft()
                for v in graph[u]:
                    if team[v] == None:
                        if team[u] == 0:
                            team[v] = 1
                            humans += 1
                        else:
                            team[v] = 0
                            robots += 1
                        queue.append(v)
            count += max(robots,humans)
    return count


n,m = map(int, input().split())
graph = [[] for _ in range(n+1)]
for _ in range(m):
    u,v = map(int, input().split())
    graph[u].append(v)
    graph[v].append(u)
print(maxNumber(graph,n))
```

```python
from collections import deque

def maxNumber(graph, n):
    football_club = [-1] * (n + 1)
    total = 0

    for i in range(1, n + 1):
        if football_club[i] == -1:
            robot_team = 1
            human_team = 0
            football_club[i] = 0
            queue = deque()
            queue.append(i)

            while queue:
                u = queue.popleft()
                for v in graph[u]:
                    if football_club[v] == -1:
                        football_club[v] = 1 - football_club[u]
                        if football_club[v] == 0:
                            robot_team += 1
                        else:
                            human_team += 1
                        queue.append(v)
            total += max(robot_team, human_team)
    return total


n, m = map(int, input().split())
graph = [[] for _ in range(n + 1)]
for _ in range(m):
    u, v = map(int, input().split())
    graph[u].append(v)
    graph[v].append(u)

print(maxNumber(graph, n))
```

```python
from collections import deque
def bfs(source, adj):
    dist = [-1] * len(adj)
    queue = deque([source])
    dist[source] = 0
    far = source
    while queue:
        u = queue.popleft()
        for v in adj[u]:
            if dist[v] == -1:
                dist[v] = dist[u] + 1
                queue.append(v)
                if dist[v] > dist[far]:
                    far = v

    return far, dist[far], dist

def find_diameter(n, edges):
    adj = [[] for i in range(n + 1)]
    for u, v in edges:
        adj[u].append(v)
        adj[v].append(u)

    a, da, dist_a = bfs(1, adj)
    b, db, dist_b = bfs(a, adj)

    return db, a, b

n = int(input())
edges = [tuple(map(int, input().split())) for i in range(n - 1)]

d, u, v = find_diameter(n, edges)
print(d)
print(u, v)
```

```python
from queue import PriorityQueue
num_words = int(input())
dictionary = [input() for _ in range(num_words)]
graph = {}
indegree = {}
alphabet_order = []
for word in dictionary:
    for char in word:
        if char not in graph:
            graph[char] = []
            indegree[char] = 0

invalid_order = False
for i in range(num_words - 1):
    word1 = dictionary[i]
    word2 = dictionary[i + 1]
    min_len = min(len(word1), len(word2))
    found_diff = False
    for j in range(min_len):
        if word1[j] != word2[j]:
            if word2[j] not in graph[word1[j]]:
                graph[word1[j]].append(word2[j])
                indegree[word2[j]] += 1
            found_diff = True
            break
    if not found_diff and len(word1) > len(word2):
        invalid_order = True
        break

if invalid_order:
    print(-1)
else:
    pq = PriorityQueue()
    for char in indegree:
        if indegree[char] == 0:
            pq.put(char)

    while not pq.empty():
        current_char = pq.get()
        alphabet_order.append(current_char)
        for neighbor in graph[current_char]:
            indegree[neighbor] -= 1
            if indegree[neighbor] == 0:
                pq.put(neighbor)

    if len(alphabet_order) == len(indegree):
        print(''.join(alphabet_order))
    else:
        print(-1)
```

```python
from collections import deque

def min_knight_moves(board_size, start_x, start_y, target_x, target_y):
    if start_x == target_x and start_y == target_y:
        return 0


    knight_moves = [(2,1),(2,-1),(1,2),(1,-2),(-1,2),(-1,-2),(-2,1),(-2,-1)]


    distance = [[-1] * (board_size + 1) for _ in range(board_size + 1)]
    distance[start_x][start_y] = 0

    queue = deque()
    queue.append((start_x, start_y))

    while queue:
        current_x, current_y = queue.popleft()
        for dx, dy in knight_moves:
            next_x = current_x + dx
            next_y = current_y + dy

            if 0 < next_x <= board_size and 0 < next_y <= board_size:
                if distance[next_x][next_y] == -1:
                    if next_x == target_x and next_y == target_y:
                        return distance[current_x][current_x]+1
                    distance[next_x][next_y] = distance[current_x][current_y] + 1
                    queue.append((next_x, next_y))

    return -1


N = int(input())
x1, y1, x2, y2 = map(int, input().split())

print(min_knight_moves(N, x1, y1, x2, y2))
```
