```python
from collections import deque

def bfs(n, adj_list, start, end):
    color = [False] * (n + 1)
    parent = [0] * (n + 1)
    queue = deque([start])
    color[start] = True

    while queue:
        u = queue.popleft()
        if u == end:
            break
        for v in adj_list[u]:
            if not color[v]:
                color[v] = True
                parent[v] = u
                queue.append(v)

    if not color[end]:
        return None


    path = []
    current = end
    while current != 0:
        path.append(current)
        if current == start:
            break
        current = parent[current]
    if path[-1] != start:
        return None
    return path[::-1]

#@####################################################################
n, m, s, d, k = map(int, input().split())
adj_list = {i: [] for i in range(1, n + 1)}
for _ in range(m):
    u, v = map(int, input().split())
    adj_list[u].append(v)
path1 = bfs(n, adj_list, s, k)
path2 = bfs(n, adj_list, k, d)

if path1 is None or path2 is None:
    print(-1)
else:
    full_path = path1 + path2[1:]
    print(len(full_path) - 1)
    print(" ".join(map(str, full_path)))
```

```python
from collections import deque

def in_bounds(r, c, R, H):
    return 0 <= r < R and 0 <= c < H

def bfs(start_r, start_c, grid, visited, R, H):
    queue = deque()
    queue.append((start_r, start_c))
    visited[start_r][start_c] = True
    diamonds = 0

    if grid[start_r][start_c] == 'D':
        diamonds += 1

    while queue:
        r, c = queue.popleft()
        for dr, dc in [(-1, 0), (1, 0), (0, -1), (0, 1)]:
            nr, nc = r + dr, c + dc
            if in_bounds(nr, nc, R, H) and not visited[nr][nc] and grid[nr][nc] != '#':
                visited[nr][nc] = True
                if grid[nr][nc] == 'D':
                    diamonds += 1
                queue.append((nr, nc))

    return diamonds

# Input
R, H = map(int, input().split())
grid = [list(input().strip()) for _ in range(R)]
visited = [[False] * H for _ in range(R)]
max_diamonds = 0

for r in range(R):
    for c in range(H):
        if grid[r][c] != '#' and not visited[r][c]:
            max_diamonds = max(max_diamonds, bfs(r, c, grid, visited, R, H))

print(max_diamonds)
```
